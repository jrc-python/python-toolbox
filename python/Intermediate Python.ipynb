### Ticks

```python
import matplotlib.pyplot as plt

year = [1961, 1981, 2001, 2021, 2041, 2061]
pop = [3.091, 4.536, 6.222, 7.874, 9.257, 10.186]

plt.plot(year, pop)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Projections')

plt.yticks([0, 2, 4, 6, 8, 10])

plt.show()
```

### Custom label ticks

```python
import matplotlib.pyplot as plt

year = [1961, 1981, 2001, 2021, 2041, 2061]
pop = [3.091, 4.536, 6.222, 7.874, 9.257, 10.186]

plt.plot(year, pop)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Projections')
plt.yticks(
    [0, 2, 4, 6, 8, 10],
    ['0', '2B', '4B', '6B', '8B', '10B']
)

plt.show()
```

### Add historical data

```python
import matplotlib.pyplot as plt

year = [1961, 1981, 2001, 2021, 2041, 2061]
pop = [3.091, 4.536, 6.222, 7.874, 9.257, 10.186]

# Add more data
year = [1800, 1850, 1900] + year
pop = [1.0, 1.262, 1.650] + pop

plt.plot(year, pop)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Projections')
plt.yticks(
    [0, 2, 4, 6, 8, 10],
    ['0', '2B', '4B', '6B', '8B', '10B']
)

plt.show()
```

### Reminder
* Square brackets []
  * Column access
    * `brics[["country", "capital"]]`
  * Row access: only through slicing
    * `brics[1:4]`
* loc (label-based)
  * Row access
    * `brics.loc[["RU", "IN", "CH"]]`
  * Column access
    * `brics.loc[:, ["country", "capital"]]`
  * Row and Column access

# Loops

NumPy Arrays

```python
import numpy as np

np_height = np.array([1.73, 1.68, 1.71, 1.89, 1.79])
np_weight = np.array([65.4, 59.2, 63.6, 88.4, 68.7])
bmi = np_weight / np_height ** 2

for val in bmi :
    print(val)
```

2D NumPy Arrays

```python
import numpy as np

np_height = np.array([1.73, 1.68, 1.71, 1.89, 1.79])
np_weight = np.array([65.4, 59.2, 63.6, 88.4, 68.7])
meas = np.array([np_height, np_weight])

for val in meas :
    print(val)
```

```python
import numpy as np

np_height = np.array([1.73, 1.68, 1.71, 1.89, 1.79])
np_weight = np.array([65.4, 59.2, 63.6, 88.4, 68.7])
meas = np.array([np_height, np_weight])

for val in np.nditer(meas) :
    print(val)
```

### Recap
* Dictionary
  * `for key, val in my_dict.items():`
* NumPy array
  * `for val in np.nditer(my_array):`

### Loop Data Structures

```python
import os

brics_location = os.getcwd() + "/python/brics.csv"
brics = pd.read_csv(brics_location, index_col = 0)

for val in brics :
    print(val)
```

### Iterrows

```python
# Slicing for better output
# The complete version would be:
#   for lab, row in brics.iterrows():
for lab, row in brics[1:3].iterrows():
    print(lab)
    print(row)
```

### Selective print

```python
import pandas as pd

brics_location = os.getcwd() + "/python/brics.csv"
brics = pd.read_csv(brics_location, index_col = 0)

for lab, row in brics.iterrows():
    print(lab + ": " + row["capital"])
```

### Add column

```python
for lab, row in brics.iterrows() :
    # - Creating Series on every iteration
    brics.loc[lab, "name_length"] = int(len(row["country"]))

print(brics)
```

### Using apply

```python
import pandas as pd

brics_location = os.getcwd() + "/python/brics.csv"
brics = pd.read_csv(brics_location, index_col = 0)

brics["name_length"] = brics["country"].apply(len)

print(brics)
```
