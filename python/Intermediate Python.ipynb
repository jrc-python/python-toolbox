# Basic plots with Matplotlib

## Data visualization

Very important in Data Analysis
* Explore data
* Report insights

### Plot

```python
import matplotlib.pyplot as plt

year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]

plt.plot(year, pop)
plt.show()
```

### Scatter plot

```python
import matplotlib.pyplot as plt

year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
plt.scatter(year, pop)
plt.show()
```

### Histogram
* Explore dataset
* Get idea about distribution

```python
import matplotlib.pyplot as plt

values = [0,0.6,1.4,1.6,2.2,2.5,2.6,3.2,3.5,3.9,4.2,6]

plt.hist(values, bins=3)
plt.show()
```

# Customization

## Data visualization

* Many options
    * Different plot types
    * Many customizations
* Choice depends on
    * Data
    * Story you want to tel

### Basic Plot

```python
import matplotlib.pyplot as plt

year = [1961, 1981, 2001, 2021, 2041, 2061]
pop = [3.091, 4.536, 6.222, 7.874, 9.257, 10.186]

plt.plot(year, pop)

plt.show()
```

### Labels

```python
import matplotlib.pyplot as plt

year = [1961, 1981, 2001, 2021, 2041, 2061]
pop = [3.091, 4.536, 6.222, 7.874, 9.257, 10.186]

plt.plot(year, pop)
plt.xlabel('Year')
plt.ylabel('Population')

plt.show()
```

### Title

```python
import matplotlib.pyplot as plt

year = [1961, 1981, 2001, 2021, 2041, 2061]
pop = [3.091, 4.536, 6.222, 7.874, 9.257, 10.186]

plt.plot(year, pop)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Projections')

plt.show()
```

### Ticks

```python
import matplotlib.pyplot as plt

year = [1961, 1981, 2001, 2021, 2041, 2061]
pop = [3.091, 4.536, 6.222, 7.874, 9.257, 10.186]

plt.plot(year, pop)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Projections')

plt.yticks([0, 2, 4, 6, 8, 10])

plt.show()
```

### Custom label ticks

```python
import matplotlib.pyplot as plt

year = [1961, 1981, 2001, 2021, 2041, 2061]
pop = [3.091, 4.536, 6.222, 7.874, 9.257, 10.186]

plt.plot(year, pop)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Projections')
plt.yticks(
    [0, 2, 4, 6, 8, 10],
    ['0', '2B', '4B', '6B', '8B', '10B']
)

plt.show()
```

### Add historical data

```python
import matplotlib.pyplot as plt

year = [1961, 1981, 2001, 2021, 2041, 2061]
pop = [3.091, 4.536, 6.222, 7.874, 9.257, 10.186]

# Add more data
year = [1800, 1850, 1900] + year
pop = [1.0, 1.262, 1.650] + pop

plt.plot(year, pop)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Projections')
plt.yticks(
    [0, 2, 4, 6, 8, 10],
    ['0', '2B', '4B', '6B', '8B', '10B']
)

plt.show()
```
